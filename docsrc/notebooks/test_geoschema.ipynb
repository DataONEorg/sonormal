```python
import json
sdoc = """{
    "@context": {
      "@vocab": "http://schema.org/",
      "geosci-time": "http://schema.geoschemas.org/contexts/temporal#"
    },
    "@type": "Dataset",
    "identifier": "http://example.net/test/01",
    "name":"Frame this",
    "dateModified": "2021-02-24T08:04:30.134Z",
    "version": 1,
    "geosci-time": {
      "@type": "time:Instant",
      "time:inTimePosition": {
        "@type": "time:ProperInterval",
        "time:hasBeginning": {
          "time:hasTRS": {
            "@id": "geosci-time:BeforePresent"
          },
          "time:numericPosition": {
            "@value": 34000000,
            "@type": "xsd:decimal"
          }
        },
        "time:hasEnd": {
          "time:hasTRS": {
            "@id": "geosci-time:BeforePresent"
          },
          "time:numericPosition": {
            "@value": 53000000,
            "@type": "xsd:decimal"
          }
        }
      }
    }
}
"""
doc = json.loads(sdoc)

frame = {
    "@context": "https://schema.org/",
    "@type": "Dataset",
    "identifier": {},
    "creator": {}
}
```

```python
import pyld.jsonld

fdoc = pyld.jsonld.frame(doc, frame=frame)

print(json.dumps(fdoc, indent=2))
```

```python
import sonormal
sonormal.installDocumentLoader()

fdoc = pyld.jsonld.frame(doc, frame=frame)

print(json.dumps(fdoc, indent=2))
```

```python
import sonormal.normalize

ndoc = sonormal.normalize.normalizeJsonld(doc)

print(json.dumps(ndoc, indent=2))
```

```python
fdoc = sonormal.normalize.frameSODataset(ndoc)
print(json.dumps(fdoc, indent=2))
```
