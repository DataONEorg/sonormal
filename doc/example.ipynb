```python
import logging
import json
import sonormal.getjsonld

logging.basicConfig(format='%(levelname)s:%(pathname)s:%(lineno)s:%(message)s', level=logging.DEBUG)

sonormal.DOCUMENT_CACHE = {}

def loghistory(resp):
    logging.debug(f"Req headers: {resp.request.headers}")
    def ph(_h):
        logging.debug(f"URL: {_h.url}")
        logging.debug(f"Status: {_h.status_code}")
        logging.debug(f"headers: {_h.headers}")

    for h in resp.history:
        ph(h)
    ph(resp)

url1 = "https://www.bco-dmo.org/dataset/3160"

#url1 = "http://pid.geoscience.gov.au/sample/AU1547827"

headers = {"Accept":"application/ld+json"}
#headers = {"Accept":"application/ld+json, application/json"}
#headers = {"Accept":"application/ld+json"}
#headers = {}
doc, res = sonormal.getjsonld.downloadJson(url1, headers=headers)

loghistory(res)

print("CACHED = ")
for k,v in sonormal.DOCUMENT_CACHE.items():
    print(k)

print(json.dumps(doc, indent=2)[:4000])
```

```python
import pyld.jsonld
edoc = pyld.jsonld.expand(doc, {"base":res.url})
doc = pyld.jsonld.compact(edoc, {"@context":"https://schema.org/"})
print("CACHED = ")
for k,v in sonormal.DOCUMENT_CACHE.items():
    print(k)
print(json.dumps(doc, indent=2)[:1000])
```
